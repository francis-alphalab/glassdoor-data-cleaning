```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("uncleaned_DS_jobs.csv")

print("Shape: " + str(df.shape))
print("Columns: " + str(df.columns.tolist()))
      
```

```python
#Explore the data
df.head(10)
```

```python
#Check data type and missing Values
print(df.dtypes)
print(df.isnull().sum())
```

```python
#Rename Columns
#Check current column names
print("Current columns:",df.columns.tolist())

#Rename columns - remove spaces and make lowercase

df.columns =df.columns.str.strip().str.lower().str.replace(" ", "_")
print("\nRename columns:")
print(df.columns.tolist())
```

```python
print(df["salary_estimate"].unique()[:5])
```

```python
df["salary_estimate"] = df["salary_estimate"].str.replace("(Glassdoor est.)","",regex=False)
print(df["salary_estimate"].unique()[:5])
                                                          
```

```python
df["salary_estimate"] = df["salary_estimate"].str.strip()
print(df["salary_estimate"].unique()[:5])
                                                          
```

```python
df["salary_estimate"] = df["salary_estimate"].str.replace("$","",regex= False)
```

```python
print(df["salary_estimate"].unique()[:5])
```

```python
df["salary_estimate"] = df["salary_estimate"].str.replace("K","000",regex=False)
print(df["salary_estimate"].unique()[:5])
```

```python
df[["min_salary", "max_salary"]] = df["salary_estimate"].str.split("-", expand=True)
print(df[["min_salary", "max_salary"]].head())
```

```python
print(df["company_name"].unique()[:5])
```

```python
df["company_name"] = df["company_name"].str.split("\n").str[0]
print(df["company_name"].head())
```

```python
print(df["job_description"].unique()[:5])
```

```python
df["job_description"] = df["job_description"].str.replace("\n", " ",regex=False)
print(df["job_description"].unique()[:5])
```

```python
df.head()
```

```python
print(df["rating"].unique())
```

```python
import numpy as np
df["rating"] = df["rating"].replace(-1, np.nan)
print(df["rating"].unique())
```

```python
print(df["size"].unique()[:5])
```

```python
df["size"] = df["size"].str.replace("to","_",regex=False)
print(df["size"].unique()[:5])
```

```python
print(df["founded"].unique())
```

```python
df["founded"] = df["founded"].replace(-1,np.nan)
print(df["founded"].unique())
```

```python
print(df["type_of_ownership"].unique())
```

```python
df["type_of_ownership"] = df["type_of_ownership"].replace("-1",np.nan)
print(df["type_of_ownership"].unique())
```

```python
df["type_of_ownership"] = df["type_of_ownership"].str.replace("-","",regex=False)
print(df["type_of_ownership"].unique())
```

```python
df["type_of_ownership"] = df["type_of_ownership"].str.replace("  "," ",regex=False)
print(df["type_of_ownership"].unique())
```

```python
print(df["industry"].unique())
```

```python
df["industry"] = df["industry"].replace("-1",np.nan)
print(df["industry"].unique())
```

```python
print(df["sector"].unique())
```

```python
df["sector"] = df["sector"].replace("-1",np.nan)
print(df["sector"].unique())
```

```python
print(df["revenue"].unique())
```

```python
df["revenue"] = df["revenue"].replace("-1",np.nan).str.replace("$","",regex=False)
print(df["revenue"].unique())
```

```python
df["revenue"] = df["revenue"].replace("Unknown / Non-Applicable",np.nan)
print(df["revenue"].unique())
```

```python
print(df["competitors"].unique())
```

```python
df["competitors"] = df["competitors"].replace("-1",np.nan)
print(df["competitors"].unique())
```

```python
df.head()
```

```python
print(df.columns.tolist())
```

```python
df = df.drop(columns=["salary_estimate"])
df["founded"] = df["founded"].astype("Int64")
```

```python
print(df.columns.tolist())
```

```python
df = df[["index","job_title", "min_salary", "max_salary", "job_description", "rating", "company_name", "location", "headquarters", "size", "founded", "type_of_ownership", "industry", "sector", "revenue", "competitors"]]
```

```python
import openpyxl
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter

df.to_excel("data cleaning.xlsx", index = False)

wb = load_workbook("data cleaning.xlsx")
ws = wb.active

for col in ws.columns:
    ws.column_dimensions[get_column_letter(col[0].column)].width = 20

wb.save("data cleaning.xlsx")
```
